```python
import pandas as pd
from os import path
import numpy as np
from random import randint
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('./data_512.csv', index_col=['timestamp', 'num'], parse_dates=['timestamp']).sort_index()
df.head()
```

```python
data_df = df[df.is_empty_den]
data_df
```

```python
seed = randint(0, 100)
np.random.seed(seed)
print("Seed is {seed}".format(seed=seed))
```

```python
train_data_df, test_data_df = train_test_split(data_df, test_size=0.1)
```

```python
len(train_data_df), len(test_data_df)
```

```python
train_data_df.to_csv("train_data_seed{seed}.csv".format(seed=seed))
```

```python
test_data_df.to_csv("test_data_seed{seed}.csv".format(seed=seed))
```
