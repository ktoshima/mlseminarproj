```python
import pandas as pd
from os import path
```

```python
df = pd.read_csv('csv/data_512.csv', index_col=['timestamp', 'num'], parse_dates=['timestamp']).sort_index()
df.head()
```

```python
def path_exists(row):
    return path.exists(row.band1_path) and path.exists(row.band3_path) and path.exists(row.hms_path)
```

```python
df['exists'] = df.apply(path_exists, axis=1)
```
